# Which one should you use? (pyplot vs matplotlib OO method?)

    * When plotting something quickly, okay to use the pyplot method
    * When plitting something mere advanced, use the OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head(10)
```

```python
# pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target");
```

```python
 # OO method
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax = ax);
# ax.set_xlim([45,100]);
```

```python
## OO method from scratch
fig,ax = plt.subplots(figsize=(10,6))

#Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                    );

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels", xlabel="Age", ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(),title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--");
```
